A. Preprocess the `pharma` press release to remove all punctuation / digits (you can use `.isalpha()` to subset)

**Resources**:

- Documentation for `.isalpha()`: https://www.w3schools.com/python/ref_string_isalpha.asp

```python
#Clean abbreviation periods - ex: U.S./N.C. vs. end of a sentence  
pattern = re.compile(r'([A-Z])\.([A-Z])\.')
abbr = pattern.sub(r'\1\2', pharma)

#Clean contractions - ex: we want "dont" or "todays", not "do" "nt" or "today" "s"
clean_apostro = re.sub(r"[’]", '', abbr)

#Remove all other punctuation and digits 
words = ''.join([char if char.isalpha() else ' ' for char in clean_apostro])

#Clean up the extra spaces
clean_pharma = ' '.join(words.split())
clean_pharma
```

```python
#Tag the string
pharma_tokens = word_tokenize(clean_pharma)
pharma_tokens_pos = pos_tag(pharma_tokens)

#Create a list of just adjectives
from collections import Counter
adjs = [one_tok[0] for one_tok in pharma_tokens_pos if one_tok[1].startswith('JJ')]
counted = Counter(adjs)
top_5 = counted.most_common(5)
print(top_5)
```

B. Write a function that takes one press release string as an input and:

- Removes named entities from each press release string (**Hint**: you may want to use `re.sub` with an or condition)
- Scores the sentiment of the entire press release using the `SentimentIntensityAnalyzer` and `polarity_scores`
- Returns the length-four (negative, positive, neutral, compound) sentiment dictionary (any order is fine)

Apply that function to each of the press releases in `doj_subset`. 

**Hints**: 

- A function + list comprehension to execute will takes about 30 seconds on a respectable local machine and about 2 mins on jhub; if it's taking a very long time, you may want to check your code for inefficiencies. If you can't fix those, for partial credit on this part/full credit on remainder, you can take a small random sample of the 717


```python
def pullnames(press_release):
    #Find the named entities
    spacy_press = nlp(press_release)
    pattern = r'\b(?:' + '|'.join(re.escape(ent.text) for ent in spacy_press.ents) + r')\b'

    #Remove named entities
    clean_press_release = re.sub(pattern, '', press_release)

    #Score the sentiment
    sent_obj = SentimentIntensityAnalyzer()
    sent_press = sent_obj.polarity_scores(clean_press_release)
    return sent_press
```

```python
sentiments = []

for release in doj_subset["contents"].tolist():
    sentiments.append(pullnames(release))

sentiments
```

## 2.1 Preprocess the data by removing stopwords, punctuation, and non-alpha words (5 points)

A. Write a function that:

- Takes in a single raw string in the `contents` column from that dataframe
- Does the following preprocessing steps:

    - Converts the words to lowercase
    - Removes stopwords, adding the custom stopwords in your code cell below to the default stopwords list
    - Only retains alpha words (so removes digits and punctuation)
    - Only retains words 4 characters or longer
    - Uses the snowball stemmer from nltk to stem

- Returns a joined preprocessed string
    
B. Use `apply` or list comprehension to execute that function and create a new column in the data called `processed_text`
    
C. Print the `id`, `contents`, and `processed_text` columns for the following press releases:

id = 16-718 (this case: https://www.seattletimes.com/nation-world/doj-miami-police-reach-settlement-in-civil-rights-case/)

id = 16-217 (this case: https://www.wlbt.com/story/32275512/three-mississippi-correctional-officers-indicted-for-inmate-assault-and-cover-up/)
    
**Resources**:

- Here's code examples for the snowball stemmer: https://www.geeksforgeeks.org/snowball-stemmer-nlp/

```python
custom_doj_stopwords = ["civil", "rights", "division", "department", "justice",
                        "office", "attorney", "district", "case", "investigation", "assistant",
                       "trial", "assistance", "assist"]

doj_subset_wscore['id'] = doj_subset_wscore['id'].str.strip()
```

```python
default_stopwords = stopwords.words('english')
all_stopwords = default_stopwords + custom_doj_stopwords
```

```python
def preprocess_str(raw_str):
    # Convert to lowercase
    text = raw_str.lower()
    
    # Tokenize and remove punctuation
    tokens = nltk.word_tokenize(text.translate(str.maketrans('', '', string.punctuation)))
    stemmer = SnowballStemmer("english")
    
    # Filter stopwords, non-alpha, and words less than 4 characters
    processed_text = [stemmer.stem(word) for word in tokens if word not in all_stopwords and word.isalpha() and len(word) >= 4]
    
    # Join tokens back into a string
    return ' '.join(processed_text)
```

```python
doj_subset_wscore['processed_text'] = doj_subset_wscore['contents'].apply(preprocess_str)
doj_subset_wscore.loc[doj_subset_wscore['id'].isin(['16-718', '16-217']), ['id', 'contents', 'processed_text']]
```

## 2.2 Create a document-term matrix from the preprocessed press releases and to explore top words (5 points)

A. Use the `create_dtm` function I provide (alternately, feel free to write your own!) and create a document-term matrix using the preprocessed press releases; make sure metadata contains the following columns: `id`, `compound` sentiment column you added, and the `topics_clean` column

**Hint**: for these, remember the pandas quantile function from pset one.  

D. Print the top 10 words for press releases in each of the three `topics_clean`

For steps B - D, to receive full credit, write a function `get_topwords` that helps you avoid duplicated code when you find top words for the different subsets of the data. There are different ways to structure it but one way is to feed it subsetted data (so data subsetted to one topic etc.) and for it to get the top words for that subset.


```python
def create_dtm(list_of_strings, metadata):
    vectorized = CountVectorizer(lowercase = True)
    dtm_sparse = vectorized.fit_transform(list_of_strings)
    dtm_dense_named = pd.DataFrame(dtm_sparse.todense(), 
        columns = vectorized.get_feature_names_out())
    dtm_dense_named_withid = pd.concat([metadata, dtm_dense_named], axis = 1)
    return(dtm_dense_named_withid)
```

```python
metadata_columns = ['id', 'compound_value', 'topics_clean']
dtm = create_dtm(doj_subset_wscore['processed_text'], doj_subset_wscore[metadata_columns])
```

```python
def get_topwords(dtm, n=10):
    word_sums = dtm.drop(columns = metadata_columns).sum().sort_values(ascending = False)
    return word_sums.head(n).index.tolist()
```

```python
pct_95 = dtm['compound_value'].quantile(0.95)
positive = dtm[dtm['compound'] >= pct_95]

topwords_positive = get_topwords(positive)
print("Top 10 words for most positive sentiment:", topwords_positive)

pct_05 = dtm['compound_value'].quantile(0.05)
negative = dtm[dtm['compound_value'] <= pct_05]

bttmwords_negative = get_topwords(negative)
print("Top 10 words for most negative sentiment:", bttmwords_negative)
```

## 2.3 Estimate a topic model using those preprocessed words (5 points)

A. Going back to the preprocessed words from part 2.3.1, estimate a topic model with 3 topics, since you want to see if the unsupervised topic models recover different themes for each of the three manually-labeled areas (civil rights; hate crimes; project safe childhood). B. After estimating the topic model, print the top 15 words in each topic.

**Hints and Resources**:

- Make sure to use the `random_state` argument within the model so that the numbering of topics does not move around between runs of your code

```python
from gensim.models import LdaModel

text_raw_tokens = [text.split() for text in doj_subset_wscore['processed_text']]

text_raw_dict = corpora.Dictionary(text_raw_tokens)
corpus_fromdict = [text_raw_dict.doc2bow(one_text) for one_text in text_raw_tokens]

lda_model = LdaModel(corpus = corpus_fromdict, id2word = text_raw_dict, num_topics = 3, passes = 10, random_state = 42)

topics = lda_model.print_topics(num_words = 15)

for topic in topics:
    words = [word.split('*')[1].strip().strip('"') for word in topic[1].split('+')]
    print(words)
```

Write an assert statement to make sure the length of the list is equal to the number of rows in the `doj_subset_wscores` dataframe

```python
l = [lda_model.get_document_topics(item, minimum_probability = 0) for item in corpus_fromdict]
assert len(l), len(doj_subset_wscore)
```

```python
rows = []

for inner_list in l:
    # Create a dictionary to store values for each topic
    topic_values = {}
    # Iterate over each tuple (topic_index, value) in the inner list
    for topic_index, value in inner_list:
        # Assign the value to the corresponding topic column
        topic_values[f'Topic_{topic_index+1}'] = value
    # Append the dictionary to the list of rows
    rows.append(topic_values)

probs = pd.DataFrame(rows)

doj_subset_wscore = pd.concat([doj_subset_wscore, probs], axis = 1)
doj_subset_wscore["top_topic"] = doj_subset_wscore[['Topic_1', 'Topic_2', 'Topic_3']].idxmax(axis=1)
```

```python
breakdown = pd.crosstab(doj_subset_wscore['topics_clean'], doj_subset_wscore['top_topic'], normalize = 'index') * 100
breakdown
```
